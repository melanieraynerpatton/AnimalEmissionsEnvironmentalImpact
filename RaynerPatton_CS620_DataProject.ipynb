# <b><font size="6">2. Data Wrangling</font></b>

## &ensp; <font size="5">2.1 Normalizing Primary Dataset</font>

```python
#remove goats, list of production values
animal_emissions = animal_emissions[(animal_emissions['Item']!='Meat of goat, fresh or chilled') & (animal_emissions['Item']!='Raw milk of goats')]
production = list(animal_emissions['Value'][animal_emissions["Element"]=='Production'])

#filter out all rows containing production values
animal_emissions = animal_emissions[animal_emissions["Element"]=='Emissions (CO2eq) (AR5)']

# remove unnecessary columns
animal_emissions = animal_emissions.drop(columns=['Domain Code','Domain','Area Code (M49)','Area','Element Code','Element','Item Code (CPC)','Year Code','Unit','Flag','Flag Description'])

#convert emissions from kgCO2eq to mtCO2e (1 kg = 0.001 metric ton)
animal_emissions["Value"] = animal_emissions["Value"].div(1000)

#rename the column that represents emissions values
animal_emissions = animal_emissions.rename(columns={"Value": "production_emissions"})

#reset the index
animal_emissions = animal_emissions.reset_index(drop=True)

#add production values back into the original df
animal_emissions['production_amount'] = production

print(animal_emissions['Item'].unique(), "\n")
print(animal_emissions.head())
```

## &ensp; <font size="5">2.1 Normalizing Secondary Datasets</font>

```python
#transpose, reset index and drop unwanted rows created from transposition
total_emissions = total_emissions.T
total_emissions = total_emissions.reset_index()
total_emissions = total_emissions.drop(columns=[1,2], index=[0,1,2])

#rename columns
total_emissions = total_emissions.rename(columns={"index": "year",0:"total_emissions"})

#select only rows with the year listed between 1961 and 2020 (reset index again)
total_emissions = total_emissions[(total_emissions["year"].astype(int)>=1961)&(total_emissions["year"].astype(int)<=2020)]
total_emissions = total_emissions.reset_index(drop=True)

print(total_emissions.head())
```

```python
pd.options.display.float_format = '{:.0f}'.format
population.columns = population.iloc[3]
population = population[population["Country Name"]=="United States"]
population = population.drop(population.columns[0:4], axis=1)
population = population.T
population.reset_index(inplace=True)
population = population.rename(columns={3:'year', 255:'population'})
population = population[(population["year"].astype(int)>=1961)&(population["year"].astype(int)<=2020)]
population.reset_index(inplace=True, drop=True)
print(population.head())
```

# <b><font size="6">3. Variable Analysis</font></b>

## &ensp; <font size="5">3.1 Data Trends</font>

```python
from functools import reduce

#emissions per item
beef = animal_emissions[animal_emissions['Item']=='Meat of cattle with the bone, fresh or chilled'].reset_index()
cows_milk = animal_emissions[animal_emissions['Item']=='Raw milk of cattle'].reset_index()
sheep = animal_emissions[animal_emissions['Item']=='Meat of sheep, fresh or chilled'].reset_index()
chicken = animal_emissions[animal_emissions['Item']=='Meat of chickens, fresh or chilled'].reset_index()
egg = animal_emissions[animal_emissions['Item']=='Hen eggs in shell, fresh'].reset_index()
pork = animal_emissions[animal_emissions['Item']=='Meat of pig with the bone, fresh or chilled'].reset_index()

#total animal emissions per year
total_animal_emissions = reduce(lambda a, b: a.add(b), [beef['production_emissions'], cows_milk['production_emissions'], sheep['production_emissions'], chicken['production_emissions'], egg['production_emissions'], pork['production_emissions']])
total_animal_production = reduce(lambda a, b: a.add(b), [beef['production_amount'], cows_milk['production_amount'], sheep['production_amount'], chicken['production_amount'], egg['production_amount'], pork['production_amount']])
```

```python
import matplotlib.pyplot as plt

#Production Emissions
plt.figure(figsize=(20,8))
plt.plot(total_emissions['year'], beef['production_emissions'], label='Beef')
plt.plot(total_emissions['year'], cows_milk['production_emissions'], label='Cow\'s Milk')
plt.plot(total_emissions['year'], sheep['production_emissions'], label='Sheep')
plt.plot(total_emissions['year'], chicken['production_emissions'], label='Chicken')
plt.plot(total_emissions['year'], egg['production_emissions'], label='Egg')
plt.plot(total_emissions['year'], pork['production_emissions'], label='Pork')
plt.title('Animal Product Emissions (by type)')
plt.xlabel('Year')
plt.ylabel('GHG Emissions (mtCO2e)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```



> Beef is the animal product with the highest emissions by far, followed by Cow's Milk, Pork, Chicken, Egg and Sheep.



```python
# Total Production
plt.figure(figsize=(20,8))
plt.plot(total_emissions['year'], beef['production_amount'], label='Beef')
plt.plot(total_emissions['year'], cows_milk['production_amount'], label='Cow\'s Milk')
plt.plot(total_emissions['year'], sheep['production_amount'], label='Sheep')
plt.plot(total_emissions['year'], chicken['production_amount'], label='Chicken')
plt.plot(total_emissions['year'], egg['production_amount'], label='Egg')
plt.plot(total_emissions['year'], pork['production_amount'], label='Pork')
plt.title('Total Animal Produciton By Type')
plt.xlabel('Year')
plt.ylabel('Amount (tons)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```



> Cow's Milk is the item with the highest total production amount, followed by chicken, pork, beef, egg and sheep.





> When observing the two graphs above, it can be determined that while beef production emits the most GHG by far when compared to other products, it does not result in the highest production amount. This gives the impression that beef production can be inefficient when it comes to emissions given the production amount. Cow's milk, which has the highest production amount, creates the second highest emission totals, though not nearly as high as beef production. It is important to understand that milk production and beef production go hand-in-hand, but that the emission and production totals are separate measurements. Pork has higher emissions than other products, but its production and emissions appear to have a similar trend. Chicken, egg and sheep production are relatively low in emissions, making it more environmentally friendly to produce than other animal products.



```python
# Total Production
plt.figure(figsize=(20,8))
plt.plot(total_emissions['year'], total_animal_production)
plt.title('Total Animal Produciton')
plt.xlabel('Year')
plt.ylabel('Amount (tons)')
plt.xticks(rotation=45)
plt.show()
```




> The total amount of animal products produced in the United States has steadily increased since 1975.




```python
#total animal emissions over time
plt.figure(figsize=(20,8))

plt.plot(total_emissions['year'], total_emissions['total_emissions'], label='Total Agriculture Emissions')
plt.plot(total_emissions['year'], total_animal_emissions, label='Animal Production Emissions')
plt.title('Total Agriculture Emissions Compared To Animal Production Emissions')
plt.xlabel('Year')
plt.ylabel('GHG Emissions (mtCO2e)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```



> The total emissions from animal production appear to follow the trend line of total agricultural green house gas emissions, as far as dips and peaks go. However, the fraction of emissions contributed by animal production have slowly declined. Even though total animal production has been increasing, their emissions have decreased. This can be attributed to external factors, such as changes in production technology, that make animal production more efficient and sustainable.



```python
#Population over time
plt.figure(figsize=(20,8))

plt.plot(total_emissions['year'], population['population'])
plt.title('Total Population Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```



> The total population of the United States has steadily increased over time.



```python
#animal production emissions compared to population
animal_prod_perp = total_animal_production/population['population']
animal_emiss_perp = total_animal_emissions/population['population']
plt.figure(figsize=(20,8))

plt.plot(total_emissions['year'], animal_emiss_perp)
plt.title('Animal Production Emissions Per Person')
plt.xlabel('Year')
plt.ylabel('Emissions (mtCO2e)')
plt.xticks(rotation=45)
plt.show()
```



> The proportion of emissions compared to the total population for each year has been on a steady decline. Despite the linear increase in population, animal production emissions have steadily decreased. This means individuals have less of an impact on production emissions, and total production emissions are not increasing like the total population.



```python
#Animal production per person
plt.figure(figsize=(20,8))
plt.plot(total_emissions['year'], animal_prod_perp)


plt.title('Animal Production Per Person')
plt.xlabel('Year')
plt.ylabel('Total Production Per Person (tons)')
plt.xticks(rotation=45)
plt.show()
```



> The dip in the beginning of the graph can be explained by stagnating production from 1960 to 1975 while the population steadily increased. Since then, the production of animal products relative to the population have exploded. This can be interpreted as overconsumption on the individual level, as well as possible overproduction. It is important to keep in mind that the United States does export some of its animal products and that there are animal products imported. This graph is good for a general representation of just how much animal products are being produced, given the population.



## &ensp; <font size="5">3.2 Variable Correlation & Statistical Significance</font>



> Correlation coefficients close to 1 imply a strong positive correlation, while correlation coefficients close to -1 imply a strong negative correlation. Statistical significance, or p-values, close to 0 indicate a strong statistical relationship. This means that these variables contribute to each other's values.



```python
import scipy

emiss_cc, emiss_p_value = scipy.stats.pearsonr(total_animal_production, total_animal_emissions)

print("Total Animal Emissions and Total Animal Production Correlation Coefficient: ", emiss_cc)
print("Total Animal Emissions and Total Animal Production statistical significance: ", emiss_p_value)
```



> Total emissions and total animal production are strongly negatively correlated and statistically significant. This means that as the total emissions decrease, animal production is likely to increase and vice versa. This is based off the trends observed in the previous graphs.



```python
pop_cc, pop_p_value = scipy.stats.pearsonr(total_animal_production, population['population'])

print("\nPopulation and Total Animal Production Correlation Coefficient: ", pop_cc)
print("Population and Total Animal Production statistical significance: ", pop_p_value)
```



> Population and total animal production are strongly positively correlated and statistically significant. This means that as population increases, total animal production also increases.



```python
popem_cc, popem_p_value = scipy.stats.pearsonr(total_animal_emissions, population['population'])

print("\nPopulation and Total Animal Production Emissions Correlation Coefficient: ", popem_cc)
print("Population and Total Animal Production Emissions statistical significance: ", popem_p_value)
```



> Population and total animal production emissions are strongly negatively correlated and are statistically significant. Based on the trends observed above, as population increases total animal production emissions decrease.





> The plots below show the relationship between the two variables, along with a linear regression line. There are also slope, intercept, and R values for each relationship. Absolute R values between 0 and 0.4 show a weak correlation, values between 0.4 and 0.8 show a moderate correlation, and values between 0.8 and 1 show a strong correlation.



```python
import numpy as np
from sklearn.linear_model import LinearRegression

x, y = total_animal_production, total_animal_emissions

slope, intercept, r, p, std_err = scipy.stats.linregress(x, y)

def slope_intercept(x):
  return slope * x + intercept

model = list(map(slope_intercept, x))

print("Slope:", slope, "\nIntercept:", intercept, "\nR:", r)

plt.scatter(x, y)
plt.plot(x, model)
plt.title('Total Animal Production vs. Animal Production Emissions')
plt.xlabel('Animal Production')
plt.ylabel('Animal Production Emissions')
plt.show()
```

# <b><font size="6">4. Conclusion</font></b>

The initial hypothesis of this project was that population, animal production, and animal production emissions would increase linearly together. This was disproven by a surprising observation that total animal emissions have actually been decreasing, despite the high, seemingly stagnant total of GHG emissions. It is common to see the statistic that 14% or so of all GHG emissions come from livestock farming. That statistic was the inspiration for this project. As it turns out, this is a global statistic. When observing the data from the United States alone, livestock production emissions make up about half of the total agricultural GHG emissions. In total, agriculture makes up about 10% of the total GHG emissions in the United States (Joiner & Toman, 2023). This means the true figure for animal production emission contribution is probably closer to 5% for the United States specifically. Below is a breakdown of how much animal production emissions have contributed to overall agriculture GHG emissions.

```python
#percent of animal production emissions compared to total agriculture ghg emissions
print([int(apm) / int(tem) for apm,tem in zip(total_animal_emissions, total_emissions['total_emissions'])])
```

Another goal of this project was to examine the impact an individual can have based on their dietary choices. Of course, choosing to be vegetarian or vegan will lower the individual's demand for animal products and decrease their carbon footprint. However, the real question is whether that individual decision will have a tangible impact on the total emissions and production in the United States. Population and animal production are almost directly linearly correlated. The more the population increases, the more production increases. These two variables directly influence each other. Based on this, if more people were to cut down on their meat consumption or cut it out all together, production would decrease. While this would save more animal lives, it wouldn't necessarily decrease emissions from production. In fact, based on the data above animal production emissions and animal production totals are negatively correlated. Despite the increase in production, emissions from production have not increased, but have actually decreased. It is logical to assume that less production will result in even fewer emissions, but as it stands animal production continues to become more efficient and sustainable. It is also worth considering the estimated 5% animal production emission contribution for the United States. This is a national statistic, and planetary warming is a global issue. This will further reduce the positive environmental impact reduced consumption in the United States offers on a global scale.
